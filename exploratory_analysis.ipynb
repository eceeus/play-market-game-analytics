```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.style.use("ggplot")
%matplotlib inline
```

```python
apps_df = pd.read_csv("data/apps_info.csv")
games_df = pd.read_csv("data/games_info.csv")
```

# kategoriye göre uygulama ve oyun dağılımı

- apps_df['clean_category']   --> apps için veri temizliği yapılmış kategori
- apps_df['categories']       --> apps için ham kategori

- games_df['categories']      --> games için veri temizliği yapılmış kategori
- games_df['cleang_category'] --> games için ham kategori

```python
num_categories = apps_df['categories'].nunique()
print(f"Toplam {num_categories} farklı kategori var.")
```

```python
apps_df['categories'].unique()
```

yukarıdaki çıktıda gördüğümüz üzere kategori isimlerimde veri temizliğine ihtiyaç var doğru kategori isminin hep en son , den sonra yer aldığını gözlemledim

```python
apps_df['clean_category'] = apps_df['categories'].apply(lambda x: x.split(',')[-1].strip())
```

```python
print("Temizlenmiş kategori sayısı:", apps_df['clean_category'].nunique())
print("Kategori isimleri:")
print(sorted(apps_df['clean_category'].unique()))
```

```python
apps_df['clean_category'].unique()
```

```python
category_counts = apps_df['clean_category'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=category_counts.index, y=category_counts.values, palette="rocket")

plt.title("Kategoriye Göre Uygulama Sayısı", fontsize=14)
plt.xlabel("Kategori")
plt.ylabel("Uygulama Sayısı")
plt.xticks(rotation=90)
sns.set_style("white")
plt.tight_layout()
plt.show()
```

```python
gnum_categories = games_df['categories'].nunique()
print(f"Toplam {gnum_categories} farklı kategori var.")
```

```python
games_df['categories'].unique()
```

```python
games_df['cleang_category'] = games_df['categories'].apply(lambda x: x.split(',')[-1].strip())
```

```python
print("Temizlenmiş kategori sayısı:", games_df['cleang_category'].nunique())
print("Kategori isimleri:")
print(sorted(games_df['cleang_category'].unique()))
```

```python
merge_map = {
    'Puzzles': 'Puzzle',
    'Sci-fi fantasy': 'Science fiction',
    'Board games': 'Board',
    'Smash & beat': 'Action',
    'Zombie': 'Horror',
    'Turn-based RPG': 'Role Playing',
    'Cafe & restaurant': 'Cooking',
    'Mine': 'Crafting',
    'Gun': 'Shooting',
    'Military': 'War',
    'Warrior': 'War',
    'Destruct': 'Action',
}
games_df['cleang_category'] = games_df['cleang_category'].replace(merge_map)
```

```python
thematic_labels = [
    'Cute', 'Peaceful', 'Immersive', 'Stylized', 'Low poly', 'Modern',
    'Realistic', 'Offline', 'Single player', 'Play Pass'
]

games_df = games_df[~games_df['cleang_category'].isin(thematic_labels)]
```

```python
games_df['cleang_category'] = games_df['cleang_category'].replace({
    'Puzzles': 'Puzzle',
    'Games': 'Game',
    'Dice': 'Board',
    'Cars': 'Car'
})
```

```python
print("Temizlenmiş son kategori sayısı:", games_df['cleang_category'].nunique())
print("Kategori isimleri:\n", sorted(games_df['cleang_category'].unique()))
```

```python
category_mapping = {
    'Eastern fantasy': 'Fantasy',
    'Medieval fantasy': 'Fantasy',
    'Sci-fi fantasy': 'Science fiction',
    'Athlete': 'Sports',
    'Stadium': 'Sports',
    'Army': 'War',
    'Military': 'War',
    'Warrior': 'War',
    'Driving': 'Racing',
    'Car': 'Racing',
    'Motorcycle': 'Racing'
}

games_df['cleang_category'] = games_df['cleang_category'].replace(category_mapping)
```

```python
category_counts = games_df['cleang_category'].value_counts()

rare_categories = category_counts[category_counts < 3].index

games_df['cleang_category'] = games_df['cleang_category'].replace(rare_categories, 'Other')
```

```python
print("Temizlenmiş kategori sayısı:", games_df['cleang_category'].nunique())
print("Kategoriler:", sorted(games_df['cleang_category'].unique()))
```

```python
category_counts = games_df['cleang_category'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(14, 6))
sns.set(style="whitegrid") 
barplot = sns.barplot(
    x=category_counts.index,
    y=category_counts.values,
    palette="crest"
)

plt.title("Kategoriye Göre Oyun Sayısı", fontsize=16)
plt.xlabel("Kategori", fontsize=12)
plt.ylabel("Oyun Sayısı", fontsize=12)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Score dağılımı analizi

```python
plt.figure(figsize=(10, 6))
sns.histplot(games_df['score'], bins=20, kde=True, color="#f877dc")
sns.set(style="whitegrid") 

plt.title("Oyun Puanı Dağılımı", fontsize=14)
plt.xlabel("Puan", fontsize=12)
plt.ylabel("Oyun Sayısı", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
bins = [0, 2, 3, 3.5, 4, 4.5, 5]
games_df['puan_segment'] = pd.cut(games_df['score'], bins)
puan_segment_counts = games_df['puan_segment'].value_counts().sort_index()

sns.barplot(x=puan_segment_counts.index.astype(str), y=puan_segment_counts.values, palette='magma')
plt.title("Puan Segmentlerine Göre Oyun Sayısı")
plt.xlabel("Puan Aralığı")
plt.ylabel("Oyun Sayısı")
plt.show()
```

```python
games_df_exploded = games_df.copy()
games_df_exploded['cleang_category'] = games_df_exploded['cleang_category'].str.split(', ')
games_df_exploded = games_df_exploded.explode('cleang_category')

category_score_means = games_df_exploded.groupby('cleang_category')['score'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x=category_score_means.values, y=category_score_means.index, palette="magma")

plt.title("Kategoriye Göre Ortalama Oyun Puanı", fontsize=14)
plt.xlabel("Ortalama Puan", fontsize=12)
plt.ylabel("Kategori", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
high_rated = games_df[games_df['score'] >= 4.5]
top_cats = high_rated['cleang_category'].value_counts().head(10)

sns.barplot(x=top_cats.values, y=top_cats.index, palette='magma')
plt.title("4.5+ Puanlı Oyunların En Yoğun Olduğu Kategoriler")
plt.xlabel("Oyun Sayısı")
plt.ylabel("Kategori")
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=games_df, x='score', y='cleang_category', orient='h', palette='rocket')
plt.title("Kategoriye Göre Puan Dağılımı (Boxplot)")
plt.xlabel("Puan")
plt.ylabel("Kategori")
plt.show()
```

# İndirme Sayısına Göre App ve Game Analizleri

```python
apps_downloads = apps_df[['downloads']].copy()
apps_downloads['type'] = 'App'
```

```python
games_downloads = games_df[['downloads']].copy()
games_downloads['type'] = 'Game'
```

```python
combined_df = pd.concat([apps_downloads, games_downloads], axis=0)
```

```python
combined_df = combined_df[combined_df['downloads'] > 0]
combined_df['log_downloads'] = np.log10(combined_df['downloads'])
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=combined_df, x='type', y='log_downloads', palette='crest')

plt.title("Uygulama ve Oyunların Ortalama İndirme Sayısı Karşılaştırması", fontsize=14)
plt.xlabel("Tür", fontsize=12)
plt.ylabel("Log10(İndirme Sayısı)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
app_category_downloads = apps_df.groupby('clean_category')['downloads'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=app_category_downloads.values, y=app_category_downloads.index, palette='crest')
plt.title('App Kategorilerine Göre Ortalama İndirme Sayısı', fontsize=14)
plt.xlabel('Ortalama İndirme Sayısı', fontsize=12)
plt.ylabel('Kategori', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=apps_df, x='score', y='downloads', alpha=0.6, color='teal')
plt.yscale('log')
plt.title('Uygulamalarda Puan ve İndirme Sayısı İlişkisi')
plt.xlabel('Puan')
plt.ylabel('İndirme Sayısı (log ölçek)')
plt.tight_layout()
plt.show()
```

```python
game_category_downloads = games_df.groupby('cleang_category')['downloads'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=game_category_downloads.values, y=game_category_downloads.index, palette='rocket')
plt.title('Game Kategorilerine Göre Ortalama İndirme Sayısı', fontsize=14)
plt.xlabel('Ortalama İndirme Sayısı', fontsize=12)
plt.ylabel('Kategori', fontsize=12)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=games_df, x='score', y='downloads', alpha=0.6, color='crimson')
plt.yscale('log') 
plt.title('Oyunlarda Puan ve İndirme Sayısı İlişkisi')
plt.xlabel('Puan')
plt.ylabel('İndirme Sayısı (log ölçek)')
plt.tight_layout()
plt.show()
```

# İndirme Segmentleriyle App/Game Dağılımı

```python
def segment_downloads(downloads):
    if downloads < 1e5:
        return 'Düşük'
    elif downloads < 1e6:
        return 'Orta'
    elif downloads < 1e7:
        return 'Yüksek'
    elif downloads < 1e8:
        return 'Çok Yüksek'
    else:
        return 'Aşırı Yüksek'

apps_df['İndirme Segmenti'] = apps_df['downloads'].apply(segment_downloads)
games_df['İndirme Segmenti'] = games_df['downloads'].apply(segment_downloads)
```

```python
apps_segment_counts = apps_df['İndirme Segmenti'].value_counts().sort_index()
games_segment_counts = games_df['İndirme Segmenti'].value_counts().sort_index()
```

```python
segment_df = pd.DataFrame({
    'Uygulamalar': apps_segment_counts,
    'Oyunlar': games_segment_counts
}).fillna(0)
```

```python
segment_df.plot(kind='bar', figsize=(10, 6), color=['#dda0dd', '#ffa54f'])

plt.title('İndirme Segmentine Göre Uygulama ve Oyun Dağılımı', fontsize=14)
plt.xlabel('İndirme Segmenti', fontsize=12)
plt.ylabel('Sayı', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.scatterplot(data=apps_df, x='downloads', y='score', alpha=0.6, color='#dda0dd')
plt.xscale('log')  
plt.title('Uygulamalar: İndirme Sayısı vs Puan')
plt.xlabel('İndirme Sayısı (log)')
plt.ylabel('Puan')
plt.ylim(0, 5.5)

plt.subplot(1, 2, 2)
sns.scatterplot(data=games_df, x='downloads', y='score', alpha=0.6, color='#ffa54f')
plt.xscale('log')
plt.title('Oyunlar: İndirme Sayısı vs Puan')
plt.xlabel('İndirme Sayısı (log)')
plt.ylabel('Puan')
plt.ylim(0, 5.5)

plt.tight_layout()
plt.show()
```
